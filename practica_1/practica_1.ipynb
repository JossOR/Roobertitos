### ¿Qué es Jupyter?

Jupyter es una aplicación web interactiva de código abierto que permite crear y compartir documentos (*notebooks*) que combinan código ejecutable, texto con formato Markdown, ecuaciones matemáticas en LaTeX y visualizaciones. ---

### ¿Qué es un archivo URDF?

Un archivo URDF (*Unified Robot Description Format*) es un formato basado en XML utilizado en ROS para describir la estructura física de un robot. Define:

- **Links (eslabones):** Partes rígidas del robot con geometría (caja, cilindro, esfera o malla), masa e inercia.
- **Joints (juntas):** Conexiones entre eslabones que especifican el tipo de movimiento (fija, rotacional, prismática, continua) y el eje de actuación.

Permite simular, visualizar y controlar el robot en herramientas como RViz y Gazebo.

---

### ¿Qué es Git y GitHub?

**Git** es un sistema de control de versiones distribuido que registra los cambios en archivos a lo largo del tiempo, permitiendo recuperar versiones anteriores y colaborar con otros desarrolladores.

**GitHub** es una plataforma en línea que aloja repositorios Git en la nube, añadiendo funcionalidades colaborativas como *pull requests*, *issues* y wikis. Es la plataforma más popular para compartir código de forma pública o privada.

---

### ¿Qué es RViz?

RViz (*ROS Visualization*) es una herramienta gráfica 3D del ecosistema ROS que visualiza en tiempo real la información generada por un robot:

- Modelos 3D descritos en archivos URDF.
- Árbol de transformaciones TF entre los marcos de referencia del robot.
- Nubes de puntos, imágenes de cámara, trayectorias y mapas.
- Comportamiento del robot durante simulación o pruebas físicas.

### 1. Creación de un modelo URDF (Robot tipo RRP)

A continuación se crea el archivo URDF de un robot serial tipo **RRP** con formas geométricas simples y dimensiones menores a 0.5 m.

**Estructura cinemática:**
- `base_link` → junta rotacional Z (q1) → `link1` (0.25 m)
- `link1` → junta rotacional Y (q2) → `link2` (0.20 m)
- `link2` → junta **prismática** Z (q3) → `link3` efector final (0.15 m)

````python
urdf_content = """<?xml version=\"1.0\"?>
<robot name=\"robot_scara\">

  <! BASE >
  <link name=\"base_link\">
    <visual>
      <geometry><cylinder radius=\"0.07\" length=\"0.05\"/></geometry>
      <origin xyz=\"0 0 0.025\" rpy=\"0 0 0\"/>
      <material name=\"gris\"><color rgba=\"0.5 0.5 0.5 1\"/></material>
    </visual>
    <collision>
      <geometry><cylinder radius=\"0.07\" length=\"0.05\"/></geometry>
      <origin xyz=\"0 0 0.025\" rpy=\"0 0 0\"/>
    </collision>
    <inertial>
      <mass value=\"1.0\"/>
      <inertia ixx=\"0.001\" ixy=\"0\" ixz=\"0\" iyy=\"0.001\" iyz=\"0\" izz=\"0.001\"/>
    </inertial>
  </link>

  <! JUNTA 1: rotacion sobre Z >
  <joint name=\"joint1\" type=\"revolute\">
    <parent link=\"base_link\"/>
    <child link=\"link1\"/>
    <origin xyz=\"0 0 0.05\" rpy=\"0 0 0\"/>
    <axis xyz=\"0 0 1\"/>
    <limit lower=\"3.14159\" upper=\"3.14159\" effort=\"10\" velocity=\"1\"/>
  </joint>

  <! ESLABON 1 >
  <link name=\"link1\">
    <visual>
      <geometry><box size=\"0.04 0.04 0.25\"/></geometry>
      <origin xyz=\"0 0 0.125\" rpy=\"0 0 0\"/>
      <material name=\"azul\"><color rgba=\"0.0 0.4 0.8 1\"/></material>
    </visual>
    <collision>
      <geometry><box size=\"0.04 0.04 0.25\"/></geometry>
      <origin xyz=\"0 0 0.125\" rpy=\"0 0 0\"/>
    </collision>
    <inertial>
      <mass value=\"0.5\"/>
      <inertia ixx=\"0.001\" ixy=\"0\" ixz=\"0\" iyy=\"0.001\" iyz=\"0\" izz=\"0.001\"/>
    </inertial>
  </link>

  <! JUNTA 2: rotacion sobre Y >
  <joint name=\"joint2\" type=\"revolute\">
    <parent link=\"link1\"/>
    <child link=\"link2\"/>
    <origin xyz=\"0 0 0.25\" rpy=\"0 0 0\"/>
    <axis xyz=\"0 1 0\"/>
    <limit lower=\"1.5708\" upper=\"1.5708\" effort=\"10\" velocity=\"1\"/>
  </joint>

  <! ESLABON 2 >
  <link name=\"link2\">
    <visual>
      <geometry><box size=\"0.04 0.04 0.20\"/></geometry>
      <origin xyz=\"0 0 0.10\" rpy=\"0 0 0\"/>
      <material name=\"verde\"><color rgba=\"0.0 0.7 0.3 1\"/></material>
    </visual>
    <collision>
      <geometry><box size=\"0.04 0.04 0.20\"/></geometry>
      <origin xyz=\"0 0 0.10\" rpy=\"0 0 0\"/>
    </collision>
    <inertial>
      <mass value=\"0.4\"/>
      <inertia ixx=\"0.001\" ixy=\"0\" ixz=\"0\" iyy=\"0.001\" iyz=\"0\" izz=\"0.001\"/>
    </inertial>
  </link>

  <! JUNTA 3: rotacion sobre Y >
  <joint name=\"joint3\" type=\"revolute\">
    <parent link=\"link2\"/>
    <child link=\"link3\"/>
    <origin xyz=\"0 0 0.20\" rpy=\"0 0 0\"/>
    <axis xyz=\"0 1 0\"/>
    <limit lower=\"1.5708\" upper=\"1.5708\" effort=\"10\" velocity=\"1\"/>
  </joint>

  <! ESLABON 3: efector final >
  <link name=\"link3\">
    <visual>
      <geometry><cylinder radius=\"0.02\" length=\"0.15\"/></geometry>
      <origin xyz=\"0 0 0.075\" rpy=\"0 0 0\"/>
      <material name=\"rojo\"><color rgba=\"0.9 0.1 0.1 1\"/></material>
    </visual>
    <collision>
      <geometry><cylinder radius=\"0.02\" length=\"0.15\"/></geometry>
      <origin xyz=\"0 0 0.075\" rpy=\"0 0 0\"/>
    </collision>
    <inertial>
      <mass value=\"0.2\"/>
      <inertia ixx=\"0.0001\" ixy=\"0\" ixz=\"0\" iyy=\"0.0001\" iyz=\"0\" izz=\"0.0001\"/>
    </inertial>
  </link>

</robot>
"""
````

**¿Qué utilidad tiene el uso de un repositorio en línea para código?**

En robótica, permite compartir modelos URDF, scripts y configuraciones de ROS de forma reproducible entre distintos equipos.

---

**¿Cuál es la convención de ángulos que utilizan los archivos URDF para los ángulos de las juntas respecto al eslabón padre?**

> Los archivos URDF utilizan la convención **RPY** (*Roll-Pitch-Yaw*) en radianes. El atributo `rpy` del elemento `<origin>` aplica rotaciones en orden: primero alrededor del eje X (roll), luego Y (pitch) y finalmente Z (yaw). El eje de movimiento de la junta (`<axis>`) se expresa en el marco del eslabón hijo, y el ángulo cero corresponde a la posición de origen definida en el `<origin>` de la junta.

---

**¿Qué utilidad tiene describir un robot en un archivo URDF?**

> El URDF centraliza toda la información cinemática y dinámica del robot (eslabones, juntas, masas, inercias y geometrías) en un solo archivo. Esto permite visualizarlo en RViz, simularlo en Gazebo, calcular automáticamente la cinemática directa e inversa con bibliotecas como KDL o MoveIt, y generar el árbol de transformaciones TF en ROS sin programación adicional.
